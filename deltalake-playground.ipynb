### UPDATE, DELETE or MERGE operations in Pyspark:
- To use these operations on delta table, we need to create a DeltaTable instance.
- Refer `forPath` and `forName` methode in [API Documentation](https://docs.delta.io/latest/api/python/index.html)

````
# Creating DeltaTable object from delta table
from delta.tables import DeltaTable

delta_tbl = DeltaTable.forName(spark, 'default.people10m')
type(delta_tbl)
````

````
# UPDATE
## Operation 3: using multiple conditions in update
delta_tbl.update(
    condition=((col("gender") == 'F') & (col("id").isin(216, 219, 232))),
    set = {"gender" : lit("Female")}
)
````

````
display(tbl.filter(col('id').isin(216, 219, 232)))
````

````
# Operation 4 and Operation 5: No updates made but a delta log file is still written in _delta_logs = 4.json, 5.json
delta_tbl.update(
    condition=((col("gender") == 'M')),
    set = {"gender" : lit("Male")}
)
````

````
# Operation 6: update made
delta_tbl.update(
    condition=((col("gender") == 'F')),
    set = {"gender" : lit("Female")}
)
````

````
tbl.filter(col("gender") == 'Female').count()
````

````
# Operation 7: Deleting records -- 30 records matching condition
delta_tbl.delete(col('birthDate') < '1955-01-01')
````

````
display(tbl.filter(col('id').isin(216, 219, 232)))
````

### History of Delta table and time travel
- Incase of delta table being written from multiple sources. It's always better to read the latest(readVersion) + 1 of delta table in a variable and continue to operate on the same dataframe.

```python
current_latest_snapshot = history_df.selectExpr("max(readVersion)").collect()[0][0] + 1
spark.read.table(f"default.people10m@v{current_latest_snapshot}")
```

````
# Returns a dataframe with all history details
history_df = delta_tbl.history()
display(history_df)
````

### Time travel
- Refer [documentation](https://docs.databricks.com/delta/history.html#delta-lake-time-travel-syntax)

````
# TIME TRAVEL ==> Also shows different ways of reading different versions of table. 
## Reading the oldest version
tbl_v0 = spark.read.format('delta').option('versionAsOf', 0).table('default.people10m') # or .load(delta_path)

# Reading version 5 - Before delete
tbl_v5 = spark.read.format('delta').option('versionAsOf', 5).load(tbl_path)

# Reading the latest version => version 7 is same as the current snapshot of table. => using @v notation, version can also be replaced with @timestamp. refer: https://docs.databricks.com/delta/history.html#delta-lake-time-travel-syntax
tbl_v7 = spark.read.table('default.people10m@v7')
````

````
tbl_v0.count(), tbl_v5.count(), tbl_v7.count()
````
